```python
import pandas as pd

# Load the dataset
students_performance = pd.read_csv('StudentsPerformance.csv')

# Display the first 5 rows of the dataset
students_performance.head()
```

```python
### The dataset displays data of students with their test scores, their background, gender, parent's education.
### The dataset looks categorical as well as numerical.
```

```python
# Check for missing values
missing_values = students_performance.isnull().sum()
print("Missing values in each column:\n", missing_values)

# Display summary statistics for numerical features
summary_statistics = students_performance.describe()
print("\nSummary statistics for numerical features:\n", summary_statistics)
```

```python
### There seems to be no missing values in this dataset. 
```

```python
# Convert categorical variables to 'category' data type
categorical_columns = ['gender', 'race/ethnicity', 'parental level of education', 'lunch', 'test preparation course']
students_performance[categorical_columns] = students_performance[categorical_columns].astype('category')

# Display the data types
data_types = students_performance.dtypes
print("Data types after conversion:\n", data_types)

# Identify which columns were converted
converted_columns = data_types[data_types == 'category'].index.tolist()
print("\nConverted columns:", converted_columns)
```

```python
### The columns that were converted are 'gender', 'race/ethnicity', 'parental level of education', 'lunch', 'test preparation course'.
```

```python
import matplotlib.pyplot as plt

# Create histograms for math, reading, and writing scores
plt.figure(figsize=(15, 5))

# Histogram for math scores
plt.subplot(1, 3, 1)
plt.hist(students_performance['math score'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Math Scores')
plt.xlabel('Math Score')
plt.ylabel('Frequency')

# Histogram for reading scores
plt.subplot(1, 3, 2)
plt.hist(students_performance['reading score'], bins=20, color='lightgreen', edgecolor='black')
plt.title('Distribution of Reading Scores')
plt.xlabel('Reading Score')
plt.ylabel('Frequency')

# Histogram for writing scores
plt.subplot(1, 3, 3)
plt.hist(students_performance['writing score'], bins=20, color='lightpink', edgecolor='black')
plt.title('Distribution of Writing Scores')
plt.xlabel('Writing Score')
plt.ylabel('Frequency')

# Show the plot
plt.tight_layout()
plt.show()
```

### Description of the distributions

Math Scores:
The distribution of math scores is approximately normal with a slight right skew. Almost displays a bell curve.
The scores range from about 0 to 100, with a higher concentration of students scoring between 60 and 80.

Reading Scores:
The distribution of reading scores appears more normally distributed.
Most students score between 60 and 90, with fewer students scoring below 50 or above 95.

Writing Scores:
The writing scores distribution is similar to the reading scores, also showing a roughly normal distribution.
Scores are concentrated between 60 and 90, with a few outliers scoring very low or very high.

```python
# Comparing scores based on gender
gender_scores = students_performance.groupby('gender').mean(numeric_only=True)[['math score', 'reading score', 'writing score']]
gender_scores.plot(kind='bar', figsize=(10, 6))
plt.title('Mean Scores by Gender')
plt.xlabel('Gender')
plt.ylabel('Mean Score')
plt.xticks(rotation=0)
plt.legend(title='Subject')
plt.show()

# Comparing scores based on race/ethnicity
race_scores = students_performance.groupby('race/ethnicity').mean(numeric_only=True)[['math score', 'reading score', 'writing score']]
race_scores.plot(kind='bar', figsize=(12, 6))
plt.title('Mean Scores by Race/Ethnicity')
plt.xlabel('Race/Ethnicity')
plt.ylabel('Mean Score')
plt.xticks(rotation=45)
plt.legend(title='Subject')
plt.show()
```

# Insights
The bar graph of Mean Scores by Gender shows female group has higher reading and writing scores than the male group. On the other hand, male group has higher math scores.

The bar graph of Mean Scores by Race/Ethnicity displays how Group E has higher scores than all other groups and group A has lower scores. Also, all the groups have one thing in common to be the reading score to be higher than other scores in most cases.
