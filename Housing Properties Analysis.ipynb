```python
# Reading the data file
import pandas as pd
ny_house = pd.read_csv("./NY-House-Dataset.csv")
ny_house
```

```python
#Calculating the average price for the houses in NY
average_price = ny_house['PRICE'].mean()
print('Average price of houses in NY is $', average_price)

#Calculating the median price for the houses in NY
median_price = ny_house['PRICE'].median()
print('Median price of houses in NY is $', median_price)

#Calculating the maximum price for the houses in NY
maximum_price = ny_house['PRICE'].max()
print('Maximum price of houses in NY is $', maximum_price)

#Calculating the minimum price for the houses in NY
minimum_price = ny_house['PRICE'].min()
print('Minimum price of houses in NY is $', minimum_price)

#Calculating the range of prices for the houses in NY
price_range = maximum_price - minimum_price
print('Range of prices of houses in NY is', price_range)
```

```python
#Calculating the average size for the houses in NY
average_propertysqrft = ny_house['PROPERTYSQFT'].mean()
print('Average size of houses in NY is', average_propertysqrft)

#Calculating the median size for the houses in NY
median_propertysqrft = ny_house['PROPERTYSQFT'].median()
print('Median size of houses in NY is', median_propertysqrft)

#Calculating the maximum size for the houses in NY
maximum_propertysqrft = ny_house['PROPERTYSQFT'].max()
print('Maximum size of houses in NY is', maximum_propertysqrft)

#Calculating the minimum size for the houses in NY
minimum_propertysqrft = ny_house['PROPERTYSQFT'].min()
print('Minimum size of houses in NY is', minimum_propertysqrft)

#Calculating the range for sizes of the houses in NY
price_propertysqrft = maximum_propertysqrft - minimum_propertysqrft
print('Range of the sizes of houses in NY is', price_propertysqrft)
```

```python
# Grouping by type
grouped_by_type = ny_house.groupby('TYPE')

# Calculate distribution statistics for prices and sizes for each property type and printing the results
price_by_type = grouped_by_type['PRICE'].describe()
print("Distribution of property prices by property type:")
print(price_by_type)

size_by_type = grouped_by_type['PROPERTYSQFT'].describe()
print("\nDistribution of property sizes by property type:")
print(size_by_type)
```

```python
# Calculating average property price for each county
county_prices = ny_house.groupby('LOCALITY')['PRICE'].mean()

# Finding the most expensive county
most_expensive_county = county_prices.idxmax()
highest_average_price = county_prices.max()
print("Most expensive county is", most_expensive_county, "with $", (highest_average_price))

# Find the least expensive county
least_expensive_county = county_prices.idxmin()
lowest_average_price = county_prices.min()
print("\nLeast expensive county is", least_expensive_county, "with $", (lowest_average_price))
```

```python
# Group the dataset by sublocality and calculate the mean for 'Property_Sqft' and 'Price' columns separately
mean_property_sqft = ny_house.groupby('SUBLOCALITY')['PROPERTYSQFT'].mean()
mean_price = ny_house.groupby('SUBLOCALITY')['PRICE'].mean()

# Combine the mean values into a DataFrame - with the help of ChatGPT
sublocality_stats = pd.DataFrame({
    'PROPERTYSQFT': mean_property_sqft,
    'PRICE': mean_price
})

# Calculate price per square foot for each sublocality
sublocality_stats['Price_per_sqft'] = sublocality_stats['PRICE'] / sublocality_stats['PROPERTYSQFT']

# Printing the results
print("Average Property Size, Average Price, and Price per Square Foot Across Different New York Sublocalities:")
print(sublocality_stats)
```
