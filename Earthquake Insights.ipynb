```python
import pandas as pd

def count_missing_values(filename):
    """
    Count the number of missing values in a CSV file.
    """
    # Read the CSV file into a DataFrame
    df = pd.read_csv(filename)
    
    # Count the missing values in each column
    missing_values_count = df.isnull().sum()
    
    # Sum the missing values counts across all columns
    total_missing_values = missing_values_count.sum()
    
    return total_missing_values

# Example usage:
filename = 'earthquakes_dataset.csv'
missing_values = count_missing_values(filename)
print(f"Number of missing values in '{filename}':", missing_values)
```

```python
import calendar
import csv
from datetime import datetime

def read_earthquake_data(filename):
    with open(filename, newline='') as csvfile:
        reader = csv.DictReader(csvfile)
        earthquakes = []
        for row in reader:
            earthquakes.append(row)
    return earthquakes

def earthquake_calendar(year, month):
    earthquakes = read_earthquake_data("earthquakes_dataset.csv")
    first_weekday, days_in_month = calendar.monthrange(year, month)
    
    day_counter = 1
    output_str = f"    {calendar.month_name[month]} {year}\n"
    output_str += "Su Mo Tu We Th Fr Sa\n"
    
    # Add initial spacing for the first week
    output_str += "   " * first_weekday
    
    while day_counter <= days_in_month:
        for weekday in range(7):
            if day_counter > days_in_month:
                break  # Break the loop if the last day is reached
            
            # Check if there was an earthquake on this day
            earthquake_on_day = False
            for earthquake in earthquakes:
                earthquake_timestamp = int(earthquake['time']) / 1000  # Convert milliseconds to seconds
                earthquake_date = datetime.utcfromtimestamp(earthquake_timestamp).date()
                if earthquake_date.year == year and earthquake_date.month == month and earthquake_date.day == day_counter:
                    earthquake_on_day = True
                    break
            
            # Print 'X' if there was no earthquake, otherwise print the day number
            if earthquake_on_day:
                output_str += f"{day_counter:2d} "
            else:
                output_str += " X "
            
            # Check if we need to move to the next line
            if (first_weekday + day_counter) % 7 == 0:
                output_str += "\n"
            
            day_counter += 1
    
    return output_str

# Example usage
print(earthquake_calendar(2018, 10))
```

```python
import calendar
import csv

def read_earthquake_data(filename):
    """
    Read earthquake data from a CSV file and return a set of dates with earthquakes.
    """
    earthquake_dates = set()
    with open(filename, 'r') as file:
        reader = csv.DictReader(file)
        for row in reader:
            time_str = row['time']
            date = time_str[:10]  # Extract only date from time string
            earthquake_dates.add(date)
    return earthquake_dates

def earthquake_calendar(year, month):
    earthquake_dates = read_earthquake_data('earthquakes_2023_global.csv')
    first_weekday, days_in_month = calendar.monthrange(year, month)
    
    day_counter = 1
    output_str = f"    {calendar.month_name[month]} {year}\n"
    output_str += "Su Mo Tu We Th Fr Sa\n"
    
    # Add initial spacing for the first week
    output_str += "   " * first_weekday
    
    while day_counter <= days_in_month:
        for weekday in range(7):
            if day_counter > days_in_month:
                break  # Break the loop if the last day is reached
            
            date = f"{year}-{month:02d}-{day_counter:02d}"
            
            # Check if there were earthquakes on this date
            if date in earthquake_dates:
                output_str += f"{day_counter:2d} "
            else:
                output_str += "X "
            
            # Check if we need to move to the next line
            if (first_weekday + day_counter) % 7 == 0:
                output_str += "\n"
            
            day_counter += 1
    
    return output_str

# Print calendar:
print(earthquake_calendar(2023, 12))
```
