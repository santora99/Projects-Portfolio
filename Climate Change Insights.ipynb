```python
# Reading the data file
import pandas as pd
climate_change_data = pd.read_csv("./climate_change_indicators.csv")
climate_change_data
```

```python
# FInding out the number of rows and columns in the dataset
climate_change_data.shape
```

```python
# Listing all column names and their data types
column_info = climate_change_data.dtypes
print("\nColumn Names and Data Types:")
print(column_info)
```

```python
# Filtering the dataset to include only the rcords for the United States
US_data = climate_change_data[climate_change_data['Country'] == 'United States']

# Count the number of records for the United States
US_record_count = len(US_data)
print("Number of records for the United States:", US_record_count)


# Calculate the average temperature change for the United States from 2000 to 2022
average_temp_change_US = US_data.loc[:, 'F2000':'F2022'].mean().mean()
print("Average temperature change for the United States from 2000 to 2022:", average_temp_change_US)
```

```python
# Select only the columns representing temperature data
temperature_columns = climate_change_data.loc[:, 'F1961':'F2022']

# Calculate the mean temperature change for each row across all years available
mean_temp_change_by_country = temperature_columns.mean(axis=1)

# Find the country with the highest average temperature change
country_highest_avg_temp_change_idx = mean_temp_change_by_country.idxmax()
highest_avg_temp_change = mean_temp_change_by_country.max()
country_highest_avg_temp_change_name = climate_change_data.loc[country_highest_avg_temp_change_idx, 'Country']
print("Country with the highest average temperature change:", country_highest_avg_temp_change_name, "with highest average temperature change:", highest_avg_temp_change)

# Filter records for the period 2010 to 2022
climate_change_2010_2022 = temperature_columns.loc[:, 'F2010':'F2022']

# Calculate the mean temperature change for each row for the period 2010 to 2022
mean_temp_change_2010_2022_by_country = climate_change_2010_2022.mean(axis=1)

# Find the top 5 countries with the highest mean temperature change for the period 2010 to 2022
top_5_countries_2010_2022_idx = mean_temp_change_2010_2022_by_country.nlargest(5).index
top_5_countries_2010_2022_names = [climate_change_data.loc[idx, 'Country'] for idx in top_5_countries_2010_2022_idx]

print("\nTop 5 countries with the highest mean temperature change from 2010 to 2022:")
print(top_5_countries_2010_2022_names)

```

```python

# Filter records for the period 2000 to 2022
data_2000_2022 = climate_change_data.loc[:, 'F2000':'F2022']

# Calculate the mean temperature change for each country for the period 2000 to 2022
mean_temp_change_2000_2022_by_country = data_2000_2022.mean(axis=1)

# Add the calculated mean temperature change as a new column to the original DataFrame
climate_change_data['Avg_Temp_Change_2000_2022'] = mean_temp_change_2000_2022_by_country

# Group the data by country and calculate the mean temperature change for each country
mean_temp_change_by_country = climate_change_data.groupby('Country')['Avg_Temp_Change_2000_2022'].mean()

print(mean_temp_change_by_country)
```

```python
# Filter records for the period from 1990 to 2022
data_1990_2022 = climate_change_data.loc[:, 'F1990':'F2022']

# Calculate the mean temperature change for each country for the period from 1990 to 2022
mean_temp_change_by_country = data_1990_2022.mean(axis=1)

grouped_mean_temp_change = mean_temp_change_by_country.groupby(climate_change_data['Country']).mean()

print(grouped_mean_temp_change)
```
